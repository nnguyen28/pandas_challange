```python
import pandas as pd
```

```python
#loading the data
school_data_load = ("C:/Users/nguye/Downloads/pandas-challenge/pandas_challange/PyCitySchool/Resources/schools_complete.csv")
student_data_load = ("C:/Users/nguye/Downloads/pandas-challenge/pandas_challange/PyCitySchool/Resources/students_complete.csv")
```

```python
#reading school and student data
school_data = pd.read_csv (school_data_load)

school_data.head()
```

```python
student_data = pd.read_csv (student_data_load)
student_data.head()
```

```python
#combinding 2 data set

complete_school_data = pd.merge(student_data, school_data, on=["school_name","school_name"], how = "left")
complete_school_data.head()
```

```python
#calculate total school
total_school = len(complete_school_data['school_name'].unique())

total_school
```

```python
#calculate total student
total_student = complete_school_data.student_name.count()
total_student
```

```python
#calculate total budget
total_budget = school_data.budget.sum()
total_budget
```

```python
#calculate average math score
average_math = complete_school_data.math_score.mean()
average_math
```

```python
#calculate average reading score
average_reading = complete_school_data.reading_score.mean()
average_reading
```

```python
#calculate the percentage passing math score (70 or greater)
percentage_math_score = complete_school_data.loc[complete_school_data['math_score'] >= 70, :].count()/complete_school_data.math_score.count()
perc_math_score = (percentage_math_score[['math_score']]*100)

perc_math_score
```

```python
#calculate the percentage passing reading score (70 or greater)
percentage_reading_score = complete_school_data.loc[complete_school_data['reading_score'] >= 70].count()/complete_school_data.reading_score.count()
perc_reading_score = (percentage_reading_score[['reading_score']]*100)

perc_reading_score
```

```python
#calculate the percentage overall passing
overall_average_passing = (average_math + average_reading)/2
overall_average_passing
```

```python
pd.DataFrame({"Total Schools":total_school , "Total Students":total_student ,"Total Budget":total_budget ,
              "Average Math Score":average_math , "Average Reading Score":average_reading , "% Passing Math":perc_math_score ,
              "% Passing Reading":perc_reading_score ,"% Overall Passing Rate":overall_average_passing})
```

```python
#calculate school types
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#calculate student per school
student_per_school = complete_school_data["school_name"].value_counts()
std_per_sch = student_per_school

std_per_sch
```

```python
#calculate budget per school
school_budget = complete_school_data.groupby(["school_name"])["budget"].mean()
budget_per_sch = school_budget

budget_per_sch
```

```python
#calculate budget per student
per_student_budget = school_budget/student_per_school
budget_per_std = per_student_budget

budget_per_std
```

```python
#calculate average math score
avg_Math_score = complete_school_data.groupby(["school_name"])["math_score"].mean()

avg_Math_score
```

```python
#calculate average reading score
avg_reading_score = complete_school_data.groupby(["school_name"])["reading_score"].mean()

avg_reading_score
```

```python
#calculate % passing math score (70 or greater)
passing_math_score = complete_school_data.loc[complete_school_data['math_score'] >= 70]
sch_pss_math_score = passing_math_score.school_name.value_counts()
perc_pss_math_score_by_sch = sch_pss_math_score/student_per_school*100

perc_pss_math_score_by_sch
```

```python
#calculate % passing reading score
passing_reading_score = complete_school_data.loc[complete_school_data['reading_score'] >= 70]
sch_pss_reading_score = passing_reading_score.school_name.value_counts()
perc_pss_reading_score_by_sch = sch_pss_reading_score/student_per_school*100

perc_pss_reading_score_by_sch
```

```python
#overall passing rate for reading and math
overall_pss_rate = (perc_pss_reading_score_by_sch+perc_pss_math_score_by_sch)/2

overall_pss_rate
```

```python
#create a datafram 
school_summary_df = pd.DataFrame({"School Type":school_types , "Total Student":std_per_sch , "Total School Budget":budget_per_sch ,
              "Per Student Budget":budget_per_std , "Average Math Score":avg_Math_score , "Average Reading Score":avg_reading_score , 
              "% Passing Math":perc_pss_math_score_by_sch , "% Passing Reading":perc_pss_reading_score_by_sch , 
              "% Overall Passing":overall_pss_rate })

school_summary_df
```

```python
# sort and display top five performing school
top_five_perf = school_summary_df.sort_values(["% Passing Math" , "% Passing Reading", "% Overall Passing"],ascending=False)

top_five_perf.head()
```

```python
#sort and display 5 worst performing school
five_worst_perf = school_summary_df.sort_values(["% Passing Math" , "% Passing Reading", "% Overall Passing"],ascending=True)

five_worst_perf.head()
```

```python
#creating bins
spending_bins = [0, 585 , 630 , 645 , 680]
spending_name = ["<$585" , "$585-630" , "$630-645" , "$645-680"]

spending_name
spending_bins
```

```python
budget_per_std.astype("float")
```

```python
school_summary_df["Spending Range(Per Student)"] = pd.cut(budget_per_std,spending_bins,labels=spending_name)

school_summary_df.head()
```

```python
#creating spending range
school_spending_range_df = school_summary_df.drop(columns=["School Type","Total Student","Total School Budget", "Per Student Budget"])
spending_range = school_spending_range_df.groupby(["Spending Range(Per Student)"])

spending_range.mean()
```

```python
#creating school sizes bins
size_bins = [0, 1000, 2000, 5000]
size_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size_names
```

```python
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Student"], size_bins , labels=size_names)

school_summary_df.head()
```

```python
#creating school sizes
school_size_df = school_summary_df.drop(columns=["School Type","Total Student","Total School Budget", "Per Student Budget"])
sch_sizes = school_size_df.groupby(["School Size"])


sch_sizes.mean()
```
